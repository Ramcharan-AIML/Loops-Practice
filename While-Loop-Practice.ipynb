```python
# Sum of numbers until user enters 0

total = 0
num = int(input("Enter the number : "))

while num != 0:
  total = total + num
  num = int(input("Enter the value except 0 : "))

print("Sum = ",total)
```

```python
# ATM Simulation

# Balance = 5000
# Keep asking withdraw amount
# Stop if amount > balance

balance = 5000

while balance > 0:
  withdraw = int(input("Enter the withdraw amount : "))

  if withdraw > balance:
    print("Insufficient Balance")
    break
  else:
    balance = balance - withdraw
    print("Current Balance : ", balance)

print("ATM session is ended")
```

```python
# Check palindrome using while


text = str(input("Enter the text : "))
i = len(text) - 1
reverse = ""

while i>=0:
    reverse = reverse + text[i]
    i = i - 1
if text == reverse:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

#IMPORTANT QUESTION FOR CODING WITH TRICKY AND CONFUSING

Screenshot 2026-02-05 122648.png

```python
seats = 350
Rejected_Bookings = 0
Total_Bookings = 0
Total_Tickets = 0

while seats >0:
    booking_seats = int(input("Enter the seats (press 0 to Exit) : "))

    if booking_seats == 0:
        break

    if booking_seats <1 or booking_seats >15:
        print("Booking Rejected (limit is 15 seats only)")
        Rejected_Bookings += 1
        continue

    if booking_seats > seats:
        print("Bookings Rejected (Insufficient seats)")
        Rejected_Bookings += 1
        continue

    for age in range(1, booking_seats+1):
        age = int(input("Enter the Age for the booking : "))

        valid_booking = True

        if age < 12:
            valid_booking = False
            break

    if valid_booking:
        print("Booking Confirmed ")
        Total_Bookings += 1
        Total_Tickets += booking_seats
        seats = seats - booking_seats
    else:
        print("Booking Rejected (Age restriction)")
        Rejected_Bookings += 1

print("Final Report :--")
print(f"Total Bookings : {Total_Bookings}")
print(f"Total Tickets Sold : {Total_Tickets}")
print(f"Rejected Bookings : {Rejected_Bookings}")
print(f"Remaning Seats : {seats}")
```
