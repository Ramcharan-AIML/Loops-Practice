```python
# FizzBuzz (Loop Version)

# From 1 to 30:
# Divisible by 3 → "Fizz"
# Divisible by 5 → "Buzz"
# Divisible by both → "FizzBuzz"

for i in range(1, 31):
  if i % 3 == 0 and i % 5 == 0:
      print("FizzBuzz")
  elif i % 3 == 0:
      print("Fizz")
  elif i % 5 == 0:
      print("Buzz")
```

```python
# Find Common Elements

# Given:
# list1 = [1, 2, 3, 4]
# list2 = [3, 4, 5, 6]

list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
common = []

for i in list1:
  for j in list2:
    if i ==j:
      common.append(i)
print(common)

```

```python
# Count Positive & Negative

# Given:
# nums = [3, -2, 7, -5, 0, 4]

# Count:
# Positive numbers
# Negative numbers

nums = [3, -2, 7, -5, 0, 4 ,-2,-1]
pos_count = 0
Neg_count = 0

for i in nums:
  if i > 0:
    pos_count = pos_count+1
  elif i < 0:
    Neg_count =Neg_count+1

print(pos_count)
print(Neg_count)
```

```python
# Prime Number Checker

# Check whether a given number is prime.

number = int(input("Enter a number : "))
count = 0

for i in range(1,number + 1):
  if number % i == 0:
    count = count +1

if count == 2:
  print("it's a prime number")
else:
  print("Not a prime number")
```

```python
# Find Second Largest Number

# Given:
# numbers = [10, 5, 8, 20, 15]

numbers = [10, 5, 8, 20, 17]
max = numbers[0]
max_2 = numbers[0]

for i in numbers:
  if i > max:
    max = i
  elif i> max_2 and i!= max:
    max_2 = i
print(max_2)
```

```python
# Check whether a string is a palindrome.

# Example:
# "madam" → Palindrome
# "python" → Not Palindrome

string = str(input("Enter the name : "))
reverse = ""

for i in string:
    reverse = i+reverse
    # print(reverse)
if string == reverse:
  print("palindrome")
else:
  print("Not a palindrome")
```

```python
# Frequency Count

# Given a string:
# text = "banana"

text = str(input("Enter the text : "))
freq = {}

for i in text:
  if i in freq:
    freq[i] = freq[i] +1
  else:
    freq[i] = 1
print(freq)

```
